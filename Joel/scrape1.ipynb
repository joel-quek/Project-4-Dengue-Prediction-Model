I want to find out all the cityname codes

```python
import re
```

```python
cityname_list =[]

for i in range(1, 64):
    element4 = wd.find_element(By.XPATH, f"/html/body/div/div/div[3]/div[1]/div[1]/div/div/ul/li[{i}]/a")
    city_code = element4.get_attribute("onclick")
    city_code = re.findall('\'([^"]*)\'', city_code)
    cityname_list.append(city_code[0])
```

## Scraping Code

```python
def weather(start_date, end_date):
    df_list = []
    for i in cityname_list:
        for j in range(int(start_date),int(end_date)+1): # range(1980,2023)
            for k in range(1,13):
                try:
                    requests.get(f"http://www.weather.gov.sg/files/dailydata/DAILYDATA_{i}_{j}{k}.csv")
                    df = pd.read_csv("http://www.weather.gov.sg/files/dailydata/DAILYDATA_{i}_{j}{k}.csv", encoding='ISO-8859-1')
                    df_list.append(df)
                except:
                    pass 
    dfs = pd.concat(df_list,ignore_index=True)
    return dfs
                
```

```python
weather(2021, 2022)
```
